```python
#W10A3
def square_sum_digits(n):
    s=0
    while(n>0):
        s+=(n%10)**2
        n//=10;
    return s;
n=0
try:
    n=int(input())
    track=set()
    while True:
        if n==1:
            print("YES")
            break
        elif n in track:
            print("NO")
            break;
        else:
            track.add(n)
            n=square_sum_digits(n)
except ValueError:
    print("n is not a positive integer")


```

```python
#W10A10
s=list(map(int,input().strip().split()))
n=s[0]
arr=[s[i] for i in range(1,n+1)]
res=0
track=set()
try:
    for i in arr:
        if i in track:
            raise ValueError
        res+=i
        track.add(i)
    print(res)
except ValueError:
    print("Mang khong hop le")
        
```

```python
#W10A11
t=int(input())
ans=[]
p=['+','*']
for _ in range(t):
    s=input()
    if s[0] in p:
        ans.append(False)
    else:
        sign=True
        for i in range(len(s)):
            if s[i] in p:
                if (i>0 and s[i-1] in p) or (i<len(s)-1 and s[i+1] in p):
                    ans.append(False)
                    sign=False
                    break
        if sign : ans.append(True)
for i in ans:
    print(i)
    
```
