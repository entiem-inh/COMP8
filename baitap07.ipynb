```python
#w5a3
n=int(input())
from math import sqrt
def snt(n):
    if ( n <= 1 ):  return False
    for i in range(2,int(sqrt(n))+1):
        if ( n % i == 0 ): return False
    return True
print(snt(n))
```

```python
#w5a4
n=int(input())
from math import sqrt
def pnum(n):
    s=1
    for i in range(2,int(sqrt(n))+1):
        if ( n % i == 0 ):
           s+= i + n/i
    if ( s == n ): return True
    return False
print(pnum(n))
```

```python
#w5a6
n=int(input())
def factorial(n):
    if ( n == 0 ): return 1
    fac=1
    for i in range( 1 , n+1 ):
        fac*=i
    return fac
print(factorial(n))
```

```python
#w5a9
n=int(input())
def tcs(n):
    s=0
    while ( n != 0 ):
      s+=int(n % 10)
      n/=10
    return s
print(tcs(n))
```

```python
#w5a10
a,b=map(str, input().split())

def kt(s1 , s2):

    map_s1_s2 = {} # Map char from s1 to s2
    map_s2_s1 = {} # Map char from s2 to s1 (to ensure one-to-one mapping)

    for i in range(len(s1)):
        char_s1 = s1[i]
        char_s2 = s2[i]

        # Check mapping from s1 to s2
        if char_s1 in map_s1_s2:
            if map_s1_s2[char_s1] != char_s2:
                return False # Inconsistent mapping for char_s1
        else:
            # If s1 char is new, ensure s2 char isn't already mapped to a different s1 char
            if char_s2 in map_s2_s1:
                return False # char_s2 is already mapped to a different char_s1
            map_s1_s2[char_s1] = char_s2
            map_s2_s1[char_s2] = char_s1

    return True

if (kt(a,b)): print("true")
else: print("false")
```
